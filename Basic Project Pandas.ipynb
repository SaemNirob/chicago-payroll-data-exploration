#### 11. What are the mean, max and min salaries?

```python
pay.info()
```

##### Annual Salary is a string, so we cannot directly calculate the mean, max, or min from it. Therefore, we first need to convert it into a float.

```python
pay['Salary'] = pay['Annual Salary'].str.replace('$','')           
```

```python
pay['Salary'] = pay['Salary'].str.replace(',', '').astype(float)
```

```python
pay.info()
```

```python
print('The Mean/Avg salary is:', '$', pay['Salary'].mean().round(2))
print('The Max salary is:', '$', pay['Salary'].max().round(2))
print('The Min salary is:', '$', pay['Salary'].min().round(2))
```

#### 24. How many people have the word 'officer' in their job title? (This is pretty tricky)

```python
def find_string(title):
    if 'officer' in title.lower():
        return True
    else:
        return False
```

```python
pay['Job Titles'].apply(find_string).sum()
```

### or

```python
sum(pay['Job Titles'].apply(find_string))
```

### or

```python
pay[pay['Job Titles'].str.contains('officer', case=False)].count()['Name']
```
